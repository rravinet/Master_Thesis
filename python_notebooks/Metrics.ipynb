```python

def generate_latex_table(selected_stocks, data):
    # Filter the data to include only the selected stocks
    filtered_data = data[data['Stock'].isin(selected_stocks)]

    # Define the horizons and metrics
    horizons = {'horizon_1': 'H1', 'horizon_5': 'H5', 'horizon_10': 'H10', 'horizon_20': 'H20'}
    metrics = ['MSE', 'RMSE']
    models = ['RG', 'RHG']

    # Create a list to store the columns
    columns = pd.MultiIndex.from_product([metrics, models], names=['Metric', 'Model'])

    # Initialize a DataFrame to store the results
    final_df = pd.DataFrame(columns=columns)

    # Populate the DataFrame
    for horizon_key, horizon_value in horizons.items():
        horizon_df = pd.DataFrame(index=selected_stocks, columns=columns)
        for stock in selected_stocks:
            stock_data = filtered_data[filtered_data['Stock'] == stock]
            for metric in metrics:
                rg_column_name = f'{metric}_realised_garch_{horizon_key}'
                rhg_column_name = f'{metric}_realised_har_garch_{horizon_key}'
                rg_value = stock_data[rg_column_name].values[0]
                rhg_value = stock_data[rhg_column_name].values[0]
                horizon_df.loc[stock, (metric, 'RG')] = rg_value
                horizon_df.loc[stock, (metric, 'RHG')] = rhg_value
        # Add the horizon column to the DataFrame
        horizon_df['Horizon'] = horizon_value
        final_df = pd.concat([final_df, horizon_df])

    # Clean up the index
    final_df.index = final_df.index.str.replace('.csv', '')

    # Add the Horizon column as the first column
    final_df.reset_index(inplace=True)
    final_df.rename(columns={'index': 'Stock'}, inplace=True)
    final_df.set_index(['Horizon', 'Stock'], inplace=True)

    # Round the numbers to 3 decimal places, excluding the 'Horizon' column
    for metric in metrics:
        for model in models:
            final_df[(metric, model)] = pd.to_numeric(final_df[(metric, model)], errors='coerce').round(3)

    # Flatten the MultiIndex columns
    final_df.columns = ['_'.join(col) for col in final_df.columns]

    # Generate LaTeX code for the table
    latex_code = """
\\begin{table}
\\centering
\\caption{Metrics for Selected Stocks Across Different Horizons}
\\label{tab:metrics_table}
\\resizebox{\\textwidth}{!}{%
\\begin{tabular}{llcccc}
\\toprule
\\textbf{Horizon} & \\textbf{Stock} & \\multicolumn{2}{c}{\\textbf{MSE}} & \\multicolumn{2}{c}{\\textbf{RMSE}} \\\\
 &  & \\textbf{RG} & \\textbf{RHG} & \\textbf{RG} & \\textbf{RHG} \\\\
\\midrule
"""
    for horizon in final_df.index.get_level_values('Horizon').unique():
        first = True
        for stock in final_df.loc[horizon].index:
            stock_data = final_df.loc[(horizon, stock)]
            if first:
                latex_code += f"\\textbf{{{horizon}}} & {stock} & {stock_data['MSE_RG']:.3f} & {stock_data['MSE_RHG']:.3f} & {stock_data['RMSE_RG']:.3f} & {stock_data['RMSE_RHG']:.3f} \\\\\n"
                first = False
            else:
                latex_code += f" & {stock} & {stock_data['MSE_RG']:.3f} & {stock_data['MSE_RHG']:.3f} & {stock_data['RMSE_RG']:.3f} & {stock_data['RMSE_RHG']:.3f} \\\\\n"
        latex_code += "\\midrule\n"
    latex_code += """
\\bottomrule
\\end{tabular}
}
\\end{table}
"""
    return latex_code

# Get the list of selected stocks
# selected_stocks = [file for file in os.listdir('/Users/raphaelravinet/Code/BSE/Thesis/final_datasets_iv') if file.endswith('.csv')]

# Assuming final_metrics is already loaded and contains all necessary data
# Generate the LaTeX table code
latex_code = generate_latex_table(selected_s, final_metrics)

# Print the LaTeX code
print(latex_code)
```

```python
def generate_latex_table(df):
    stocks = df['Stock'].unique()
    horizons = df['Horizon'].unique()
    
    latex_code = """
\\begin{table}[ht]
\\centering
\\begin{tabular}{|l|cc|cc|cc|cc|}
\\hline
 & \\multicolumn{2}{c|}{\\textbf{No\\_of\\_violations}} & \\multicolumn{2}{c|}{\\textbf{Violation\\_rate}} & \\multicolumn{2}{c|}{\\textbf{UC\\_p\\_value}} & \\multicolumn{2}{c|}{\\textbf{DQ\\_p\\_value}} \\\\
\\textbf{Stock} & \\textbf{RH} & \\textbf{RHG} & \\textbf{RH} & \\textbf{RHG} & \\textbf{RH} & \\textbf{RHG} & \\textbf{RH} & \\textbf{RHG} \\\\
\\hline
"""

    for horizon in horizons:
        latex_code += f"\\multicolumn{{9}}{{|l|}}{{\\textbf{{{horizon}}}}} \\\\ \\hline\n"
        for stock in stocks:
            stock_data = df[(df['Stock'] == stock) & (df['Horizon'] == horizon)]
            if not stock_data.empty:
                rh_data = stock_data[stock_data['Model'] == 'RG']
                rhg_data = stock_data[stock_data['Model'] == 'RHG']
                
                rh_values = rh_data[['No_of_violations', 'Violation_rate', 'UC_p_value', 'DQ_p_value']].values.flatten() if not rh_data.empty else ['']*4
                rhg_values = rhg_data[['No_of_violations', 'Violation_rate', 'UC_p_value', 'DQ_p_value']].values.flatten() if not rhg_data.empty else ['']*4
                
                rh_values = [f"{x:.3f}" if isinstance(x, float) else x for x in rh_values]
                rhg_values = [f"{x:.3f}" if isinstance(x, float) else x for x in rhg_values]
                
                latex_code += f"{stock} & {' & '.join(map(str, rh_values))} & {' & '.join(map(str, rhg_values))} \\\\\n"
        
    latex_code += "\\hline\n\\end{tabular}\n\\caption{Caption}\n\\label{tab:label}\n\\end{table}"
    
    return latex_code

# Generate LaTeX code for the table
latex_table = generate_latex_table(es_var)
print(latex_table)
```

```python
import pandas as pd
import os
from scipy.stats import skew, kurtosis

# Define the directory containing the CSV files
directory = '/Users/raphaelravinet/Code/BSE/Thesis/final_datasets copy 2'

# Define the metrics to compute
metrics = ['Obs', 'Mean', 'Std', 'Median', 'Skew', 'Kurt', 'Q1', 'Q3']

# Function to compute statistics
def compute_stats(data, column, metrics):
    stats = {}
    if 'Obs' in metrics:
        stats['Obs'] = len(data)
    if 'Mean' in metrics:
        stats['Mean'] = data[column].mean()
    if 'Std' in metrics:
        stats['Std'] = data[column].std()
    if 'Median' in metrics:
        stats['Median'] = data[column].median()
    if 'Skew' in metrics:
        stats['Skew'] = skew(data[column])
    if 'Kurt' in metrics:
        stats['Kurt'] = kurtosis(data[column], fisher=False)
    if 'Q1' in metrics:
        stats['Q1'] = data[column].quantile(0.25)
    if 'Q3' in metrics:
        stats['Q3'] = data[column].quantile(0.75)
    return stats

# Initialize a list to hold all results
all_results = []

# Process each file in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        filepath = os.path.join(directory, filename)
        data = pd.read_csv(filepath)
        
        # Multiply 'returns' and 'RM' by 100
        if 'returns' in data.columns:
            data['returns'] = data['returns'] * 100
        if 'RM' in data.columns:
            data['RM'] = data['RM'] * 100
        
        # Compute statistics for 'returns' and 'RM'
        returns_stats = compute_stats(data, 'returns', metrics) if 'returns' in data.columns else {}
        rm_stats = compute_stats(data, 'RM', metrics) if 'RM' in data.columns else {}
        
        # Combine statistics into a DataFrame
        stats_df = pd.DataFrame({
            'Metric': metrics,
            'Returns': [returns_stats.get(metric, 'N/A') for metric in metrics],
            'RM': [rm_stats.get(metric, 'N/A') for metric in metrics]
        })
        
        # Print the computed statistics for each file
        print(f'Statistics for {filename}:')
        print(stats_df)
        print('\n')
        
        # Append to all results
        all_results.append((filename, stats_df))
```
